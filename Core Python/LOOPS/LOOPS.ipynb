# printing patterns using for loops

## 1.

```python
num = 11
for i in range(1,num):
    print('*' * i)
```

## 2. 

```python
num = 10
for i in range(num, 0,-1):
    print('*' * i)
```

## 3. 

```python
num = 10
for i in range(1,num+1):
    print('*' * i)
for j in range(num-1,0,-1):
    print('*' * j)
```

## 4. 

```python
num_ = 6
print('*')
for i in range(1,num_):
    print('*' + ' ' * i + '*')
for j in range(num_-2,0,-1):
    print('*' + ' ' * j + '*')
print('*')
```

## 5.

```python

num = 10
print('*' * num)
for i in range(num):
    print('*' + ' ' * (num-2) + '*')
print('*' * num)
```

## 6. 

```python
num = 10
print('* ' * num)
for i in range(num):
    print('*' + ' ' * (num * 2 - 3) + '*')
print('* ' * num)
```
