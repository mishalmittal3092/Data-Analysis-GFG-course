```python
import json
import time
```

```python
fd1 = open("bakery inventory.json", "r")
js1 = fd1.read()
record = json.loads(js1)
fd1.close()

print("-" * 15 , "MENU", "-" * 15 , "\n")

for key in record.keys():
    print(key,":", record[key]["name"],"|",record[key]["price"],"|", record[key]["quantity"])

    
print("-"*30 )
ui_name = input("enter your name")
ui_phone = str(input("enter the phone number"))
ui_mail =  str(input("enter your mail id"))
ui_prid = input("enter the product id   :")
ui_quan = int(input("enter the quantity     :"))

print("-"*30 )

for i in record.keys():
    if i == ui_prid:
        if ui_quan <= record[i]["quantity"]:
        
            print("name           : " , record[i]["name"])
            print("price          : " , record[i]["price"])
            print("quantity       : " , ui_quan)
            print("-"*30 )
            print("billing amount : " , ui_quan * record[i]["price"])

            record[i]["quantity"] = record[i]["quantity"] - ui_quan
            
            fd_sales = open("sales.txt", "a")
            sale = ui_name + "," + ui_phone + "," + ui_mail + "," + ui_prid + "," + record[i]["name"] + "," + str(record[i]["price"]) + "," + str(ui_quan) + ","+ str(ui_quan * record[i]["price"]) + "," + time.ctime() + "\n"
            fd_sales.write(sale)
            fd_sales.close()
        else:
            print("we dont have enough quantity of demand")
            print("we only have", record[i]["quantity"])
            print("do you want to buy the remaining products")
            choice = input("enter Y/N")

            if choice =="Y" or choice == "y" and record[i]["quantity"] >0:
                print("name           : " , record[i]["name"])
                print("price          : " , record[i]["price"])
                print("quantity       : " , record[i]["quantity"])
                
                print("-"*30 )
                
                print("billing amount : " , record[i]["quantity"] * record[i]["price"])

                fd_sales = open("sales.txt", "a")
                sale = ui_name + "," + ui_phone + "," + ui_mail + "," + ui_prid + "," + record[i]["name"] + "," + str(record[i]["price"]) + "," + str(record[i]["quantity"]) + ","+ str(record[i]["quantity"] * record[i]["price"]) + "," + time.ctime() + "\n"
                fd_sales.write(sale)
                fd_sales.close()
                
                record[i]["quantity"] = record[i]["quantity"] - record[i]["quantity"]
            else: 
                print("thanks")

                
print("-"*50 )

js2 = json.dumps(record)
fd2 = open("bakery inventory.json" , "w")
fd2.write(js2)
fd2.close()


print("THE FINAL INVENTORY IS ")
fd3 = open("bakery inventory.json", "r")
js3 = fd3.read()
record_final = json.loads(js3)
fd3.close()

for key in record_final.keys():
    print(key,":", record_final[key]["name"],"|",record_final[key]["price"],"|", record_final[key]["quantity"])
print('-' * 30)
print("the final sales table is")
fd_check = open("sales.txt","r")
rec = fd_check.read()
fd_check.close()

print(rec)



```
