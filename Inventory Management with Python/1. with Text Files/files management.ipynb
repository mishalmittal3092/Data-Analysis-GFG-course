```python
import time 

fd_inven = open('inventory.txt','r')
txt = fd_inven.read()
inven = txt.split("\n")
product = []

for i in inven:
    product.append( i.split(','))
    
ui_name = input("enter your name")
ui_phone = input("enter your phone number")
ui_mail = input("enter your mail")
ui_prod_id = input("enter a product id: ")
ui_prod_quan = input("enter the quantity you want to buy: ")

for j in product:
    if j[0] == ui_prod_id:
        if int(j[4]) >= int(ui_prod_quan):
            print("product name   : ", j[1])
            print("product size   : ", j[2])
            print("product rate   : ", j[3])
            print("*"* 50)
            print("billing amount : ", int(j[3]) * int(ui_prod_quan))
            j[4] = int(j[4]) - int(ui_prod_quan)
            print("left stock     : " , j[4])

            fd_sales = open('sales.txt','a')
            sales_details = ui_name + ',' + ui_phone +',' + ui_mail + ',' + ui_prod_id + ',' + ui_prod_quan + ',' + str(int(j[3]) * int(ui_prod_quan)) + ',' + time.ctime() + '\n'
            fd_sales.write(sales_details)
            fd_sales.close()
        else:
            print("we dont have enough quantity")
            print("we only have", j[4], "quantity")
            print("do you want to buy them")

            choice = input("enter Y/N")

            if choice =='Y' or choice == 'y' and int(j[4]) >0:
                print("product name   : ", j[1])
                print("product size   : ", j[2])
                print("product rate   : ", j[3])
                print("*"* 50)
                print("billing amount : ", int(j[3]) * int(j[4]))
                
                fd_sales1 = open('sales.txt','a')
                sales_details = ui_name + ',' + ui_phone +',' + ui_mail + ',' + ui_prod_id + ',' + j[4] + ',' + str(int(j[3]) * int(j[4])) + ',' + time.ctime() + '\n'
                fd_sales1.write(sales_details)
                fd_sales1.close()
            
                j[4] = int(j[4]) - int(j[4])
                print("left stock     : " , j[4])
            else:
                print("thanks")
                
updated = []
for item in product:
    updated.append(str(item[0]) + ',' + str(item[1]) + ',' + str(item[2]) + ',' + str(item[3]) + ',' + str(item[4]) + '\n')
updated[-1] = updated[-1][:-1]
fd_inven.close()

fd_update = open ('inventory.txt', 'w')
for i in updated:
    fd_update.write(i)
    
fd_update.close()

print('-'* 50)

print("inventory updated and sales closed the final inventory list is")
fd_check = open('inventory.txt', 'r')
print(fd_check.read())
fd_check.close()

print('*'* 50)

print("the final sales table is")
fd_sale_check = open('sales.txt', 'r')
print(fd_sale_check.read())
fd_sale_check.close()

```
