# Data Cleaning

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.impute import SimpleImputer as si
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df.isna().sum()
```

## Dealing with NULL values
### Using Simple Imputer with numerical columns

```python
imputer = si(missing_values = np.nan , strategy = 'mean')
imputer.fit(df.iloc[:,13:14].values)

df.iloc[:,13:14] = imputer.transform(df.iloc[:,13:14].values)
```

### Using Simple Imputer with Categorical columns

```python
imputer = si(missing_values = np.nan , strategy = 'most_frequent')
imputer.fit(df.iloc[:,12:13].values)

df.iloc[:,12:13] = imputer.transform(df.iloc[:,12:13].values)
```

### dropping all other rows with NULL values


```python
df = df.dropna()
```

```python
df.isna().sum()
```
