### 1. working with pandas Series

#### **a) Creating Series**
 
 

##### Series through list
 
 

```python
sr1 = pd.Series([1,2,3,4,5,6])
sr1
```

##### Series through Numpy Array
 
 

```python
arr = np.array([1,2,3,4,5,6])
sr2 = pd.Series(arr)
sr2
```

##### Setting up our own index
 
 

```python
lst = [10,20,35,25,45,26]
sr3 = pd.Series(data = lst, index = ['mayur','shaurya','rajan','anjaney','yash','anant'])
sr3
```

##### Series through dictionary
 
 

```python
dict1 = {'mayur': 10, 'shaurya':20, 'rajan' : 35, 'anjaney' : 25, 'yash' : 45, 'anant': 26}
sr4 = pd.Series(dict1)
sr4
```

##### Using repeat function along with creating a Series
 
 

```python
sr5 = pd.Series([10,5]).repeat([5,2])
sr5
```

##### Using reset_index for proper indexing 

```python
sr6 = pd.Series([10,5]).repeat([5,2]).reset_index(drop = True)
sr6
```

##### Accessing data from Series

```python
print(sr6[1])
```

#### b) Aggregate function on Pandas Series


```python
sr1.agg([min,max,sum])
```

#### c) Series Absolute Function


```python
sr7 = pd.Series([1,2,-3,-4,5,6])
sr7.abs()
```

#### e) Astype Function


```python
sr6.astype('float')
```

#### f) Between Functions


```python
sr6.between(5,10)
```

#### g) All strings functions can be used to extract or modify texts in a series

```python
sr = pd.Series(['    mayur dange  ','    shaurya gupta'    ,'    rajan gupta   ','anjaney tiwari','yash gupta   ','anant tiwari   '])

sr.str.upper()
```

```python
sr.str.lower()
```

##### Len function
 
 

```python
for i in sr:
    print(i, ':',len(i))
```

##### Strip Function
 
 

```python
sr_new = sr.str.strip()
sr_new
```

##### Split Function
 
 

```python
sr_new.str.split(' ')
```

##### Contains Function
 
 

```python
sr_new.str.contains('dange')
```

##### Replace Function
 
 

```python
sr_new.str.replace('anant','harshit')
```

##### Count Function
 


```python
sr.str.count('a')
```

 ##### Startswith Function
 


```python
sr_new.str.startswith('a')
```

##### Endswith function

```python
sr_new.str.endswith('gupta')
```

 ##### Find Finction


```python
sr_new.str.find('gupta')
```

```python
df.loc[1:3, 'one':'three']
```

```python
df.loc[[1,2],['two','four']]
```

 ##### Basic Iloc Operations

```python
df.iloc[1:3, 0:2]
```

```python
df.iloc[[0,1],[1,2]]
```

##### Slicing Using Conditions

```python
df.loc[df['two'] > 20,['two']]
```

#### k) Pipe Functions:


```python
df
```

```python
def add_(i,j):
    return i+j

df.pipe(add_,10)
```

```python
def mean(i):
    return i.mean()
def square(i):
    return i**2
print(df.pipe(mean))
print(df.pipe(mean).pipe(square))
```

#### o) Groupby Functions


```python
cricket = {'Team'   : ['India', 'India', 'Australia', 'Australia', 'SA', 'SA', 'SA', 'SA', 'NZ', 'NZ', 'NZ', 'India'],
           'Rank'   : [2, 3, 1,2, 3,4 ,1 ,1,2 , 4,1,2],
           'Year'   : [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
           'Points' : [876,801,891,815,776,784,834,824,758,691,883,782]}

df_cric = pd.DataFrame(cricket)
df_cric
```

```python
df_cric.groupby('Team').groups
```

```python
df_cric.groupby('Team').get_group('Australia')
```

 ##### Adding Statistical Computation on groupby


```python
df_cric.groupby('Team').sum()['Points']
```

```python
df_cric.groupby('Team').sum()['Points'].sort_values(ascending = False)
```

 ##### Using Filter Function with Groupby

```python
df_cric.groupby('Team').filter(lambda x: len(x) >= 3).sort_values(by = 'Team', ascending = True)
```
