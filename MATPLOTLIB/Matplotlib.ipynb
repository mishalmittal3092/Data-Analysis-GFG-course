### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
```

### 1. Creating Sine & Cosine graphs

```python
x = np.linspace(0,4 * np.pi,101)

sine = np.sin(x)
cosi = np.cos(x)

plt.plot(x, sine, label = 'sine')
plt.plot(x,cosi,label = 'cosine')

plt.legend(loc = 'upper right')
plt.title('sine & cosine waves')

plt.show()
```

```python
x = np.linspace(0,4 * np.pi,101)

sine = np.sin(x)
cosi = np.cos(x)

plt.plot(x, sine, '--')
plt.plot(x,cosi, '*')

plt.legend(loc = 'upper right')
plt.title('sine & cosine waves')

plt.show()
```

### 2. setting different size

```python
plt.figure(figsize = (15, 6))

x = np.linspace(0, 4* np.pi, 101)

sine = np.sin(x)
cosi = np.cos(x)

plt.plot(x, sine, '--')
plt.plot(x,cosi, '*')

plt.legend(loc = 'upper right')
plt.title('sine & cosine waves')

plt.show()
```

### 3. Setting style to graphs

```python
print(plt.style.available)
```

```python
plt.figure(figsize = (15, 6))
plt.style.use('bmh')


plt.plot(x,sine,label = 'sine')
plt.plot(x,cosi, label = 'cosine')

plt.legend(loc = 'upper right')
plt.title('sine & cosine waves')
plt.show()
```

### 4. Saving the graphs in local system

```python
plt.figure(figsize = (20, 6))
plt.style.use('classic')

x = np.linspace(0, 4* np.pi, 101)

sine = np.sin(x)
cosi = np.cos(x)


fig = plt.figure()
plt.plot(x,sine,label = 'sine')
plt.plot(x,cosi, label = 'cosine')

plt.legend(loc = 'upper right')
plt.title('sine & cosine waves')
plt.show()

fig.savefig('sine & cosine wave.png')
```

### 6. Line plots

#### 6.1  basic Line Plot

```python
plt.style.use('bmh')
plt.figure(figsize = (10,5))

x = np.linspace(0, 4* np.pi, 101)

tan = np.tan(x)
plt.title('Tan wave')
plt.plot(x,tan)
plt.show()
```

#### 6.4 Adjusting the Plot Axes

```python
plt.figure(figsize = (11, 6))
plt.style.use('bmh')

x = np.linspace(-4*np.pi, 4*np.pi,1001)

sine = np.sin(x)
cosi = np.cos(x)

plt.axis([-1,12,-1.1,1.1])
plt.plot(x,sine,label = 'sine')
plt.plot(x,cosi, label = 'cosine')


plt.legend(loc = 'upper right')
plt.title('sine & cosine waves')
plt.show()
```

#### 6.5 Set

```python
x = np.linspace(-4*np.pi, 4*np.pi,1001)
plt.figure(figsize = (11,6))
plt.style.use('bmh')
sine = np.sin(x)
cosi = np.cos(x)

fig = plt.axes()
fig.plot(x,sine,label = 'sine')
fig.plot(x,cosi, label = 'cosine')

fig.set(xlim = (-1,12), ylim = (-1.1,1.1), xlabel = 'x', ylabel = 'sine', title = 'sine & cosine waves')

fig.legend(loc = 'upper right')
```

### 7 Scatter Plots



```python
n = 1000
plt.figure(figsize = (8,7))
plt.style.use('Solarize_Light2')

x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x,y, c = 'red', edgecolor = 'black', alpha = 0.7, s = 40)
plt.title('Scatter Plot')

plt.show
```

#### 7.1 Multiple Scatter Plots

```python
n= 500
x = np.random.rand(n)
y = np.random.rand(n)

plt.figure(figsize = (9,7))

plt.scatter(x,y, c = 'red', edgecolor = 'black', alpha = 0.6, label = 'random 1')
plt.scatter(x+0.5, y, c = 'violet', edgecolor = 'black', alpha = 0.6, label = 'random2')
plt.scatter(x+0.5, y + 0.5, c = 'blue', edgecolor = 'black', alpha = 0.6, label = 'random2')
plt.scatter(x, y + 0.5, c = 'green', edgecolor = 'black', alpha = 0.6, label = 'random2')

plt.title('Multiple Scatter Plot')
plt.legend(loc = 'upper right')

plt.show()
```

#### 7.2 changing size and style of markers

```python
plt.figure(figsize = (8,7))
rng = np.random.RandomState(0)

x = rng.randn(200)
y = rng.randn(200)

sizes = rng.rand(200) *800
colors = rng.rand(200)

plt.scatter(x,y, c = colors, s = sizes, cmap = 'magma' , alpha = 0.6, edgecolors = 'black')

plt.title('Random Scatter Plot with different sizes and colors')

plt.colorbar()
plt.show()
```

### 9. Pie Charts




```python
plt.figure(figsize = (7,7))
cars = ['BMW', 'PORSCHE', 'JAGUAR', 'DEFENDER', 'CRETA', 'FORTUNER', 'BENZ']
sales = [435343,464643,2362456,565455,765754,626265,526254]

plt.pie(sales , labels = cars)
plt.show()
```

#### 9.1 Basic Operations

```python
plt.figure(figsize = (7,7))
cars = ['BMW', 'PORSCHE', 'JAGUAR', 'DEFENDER', 'CRETA', 'FORTUNER', 'BENZ']
sales = [435343,464643,2362456,565455,765754,626265,526254]
explodes = [.05,.1,.1,.1,.1,.1,.1]

plt.pie(sales , labels = cars, explode = explodes )
plt.show()
```

```python
plt.figure(figsize = (7,7))
cars = ['BMW', 'PORSCHE', 'JAGUAR', 'DEFENDER', 'CRETA', 'FORTUNER', 'BENZ']
sales = [435343,464643,2362456,565455,765754,626265,526254]
explodes = [0,0,0,0,0,.1,0]

plt.pie(sales , labels = cars, explode = explodes, shadow = True,autopct = '%.2f' )
plt.show()
```

```python
plt.figure(figsize = (7,7))
cars = ['BMW', 'PORSCHE', 'JAGUAR', 'DEFENDER', 'CRETA', 'FORTUNER', 'BENZ']
sales = [435343,464643,2362456,565455,765754,626265,526254]
explodes = [0,0,0,0,0,.1,0]

plt.pie(sales , labels = cars, explode = explodes, shadow = True,autopct = '%.2f',
       wedgeprops = {'edgecolor': 'black', 'linewidth': 1 })
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.style.use('bmh')
cars = ['BMW', 'PORSCHE', 'JAGUAR', 'DEFENDER', 'CRETA', 'FORTUNER', 'BENZ']
sales = [435343,464643,2362456,565455,765754,626265,526254]
explodes = [0,0,0,0,0,.1,0]

plt.pie(sales , labels = cars, explode = explodes, shadow = True,autopct = '%.2f%%',
        wedgeprops = {'edgecolor': 'black', 'linewidth': 1 , 'antialiased' : True}, textprops = {'color' : 'white'} )
plt.show()
```

### 10. Multiple Subplots

```python
plt.figure(figsize = (14,9))
plt.style.use('bmh')
x = np.linspace(0, 4*np.pi, 101)

plt.subplot(2,2,1)
sine = np.sin(x)
plt.plot(x,sine)
plt.title('sine curve')


plt.subplot(2,2,2)
cosi = np.cos(x)
plt.plot(x,cosi)
plt.title('cosine curve')

plt.subplot(2,2,3)
tang = np.tan(x)
plt.plot(x,tang)
plt.title('tangent curve')

plt.subplot(2,2,4)
x = np.random.randn(100)
plt.hist(x)
plt.title('Random Histogram')


plt.tight_layout()
plt.show()
```

```python
for i in range(1,7):
    plt.subplot(2,3,i)
    
    plt.text(0.5,0.5,str((2,3,i)), ha = 'center', fontsize = 18)

plt.tight_layout()
plt.show()
```

### 11. 3D Plots

```python
ax = plt.axes(projection = '3d')
```

```python
x = np.linspace(0,4*np.pi, 10000)

ax = plt.axes(projection = '3d')
xa = np.sin(x)
y = np.cos(x)
z = np.linspace(0,35,10000)

ax.plot3D(xa,y,z,'blue')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

x = np.random.randn(200)
y = np.random.randn(200)
z = np.random.randn(200)

ax.scatter(x,y,z, c = 'red', s = 500, alpha = 0.6, marker = '.')

plt.tight_layout()
plt.show()
```
